```python
from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

# Load data

```python
iris = load_iris() #
X, y = iris.data, iris.target

X_train, X_test, y_train,y_test = train_test_split(X, y, test_size = 0.2, shuffle=True)
```

```python
print("shape of X_train :", X_train.shape)
print("shape of X_test  :", X_test.shape)
```

```python
clf = tree.DecisionTreeClassifier(criterion = "entropy")
clf = clf.fit(X_train, y_train)
```

# Draw

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file="./img/Iris_Decision_Tree_sklean.dot")
graph = graphviz.Source(dot_data)
#graph.render("iris")
dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names= iris.feature_names,  
                         class_names=  iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)


graph = graphviz.Source(dot_data)
graph 
```

# Test: decision tree with criterion Entropy

```python
y_hat = clf.predict(X_test, check_input=True)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_hat)
```

# Decision tree with criterion Gini Index

```python
clf_gini = tree.DecisionTreeClassifier(criterion = "gini")
clf_gini = clf_gini.fit(X_train, y_train)
```

```python
dot_data = tree.export_graphviz(clf_gini, out_file="./img/Iris_Decision_Tree_sklean_gini.dot")
graph = graphviz.Source(dot_data)
#graph.render("iris")
dot_data = tree.export_graphviz(clf_gini, out_file=None, 
                         feature_names= iris.feature_names,  
                         class_names=  iris.target_names,  
                         filled=True, rounded=True,  
                         special_characters=True)


graph = graphviz.Source(dot_data)
graph 
```

# Compare Gini index vs Information Gain

```python
y_hat_gini = clf_gini.predict(X_test, check_input=True)
accuracy_score(y_test, y_hat)
```
